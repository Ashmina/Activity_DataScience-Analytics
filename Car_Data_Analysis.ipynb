**Objective:** To analyze the dataset 'cars_data.csv' and do the following operations:
1. Read the dataset to the python environment.

3. Plot a bar graph of male vs female buyers participated in the sales.

5. Find the least 5 cars based on their Resell price.


```python
#importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

### 1. Reading the dataset to python environment

```python
#reading the dataset
car_data = pd.read_csv('cars_data.csv')
```

```python
#displaying first 5 rows
car_data.head()
```

### 2. Checking for null values present in the dataset

```python
car_data.isnull().sum()
```

There are no missing(null) values in the cars_data dataset.

### 3. Bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(6,6))
m_f_plot = sns.countplot(x="Buyer Gender",data=car_data,palette=["#FFC0CB","#ADD8E6"])
plt.title("Male Vs Female Buyers",fontsize=20)
plt.xlabel("Buyer Gender",fontsize=15)
plt.ylabel("Count",fontsize=15)
#labelling the bars with the data value
plt.bar_label(m_f_plot.containers[0],fontsize=12)
plt.show()
```

The number of Female buyers are slightly higher.

### 4.Find the top 5 cars based on their sales price.

```python
#function to give a background colour
def highlight_cols(x):
    df=x.copy()
    df.loc[:, :] = 'background-color: #EEEEE0'
    df[['Make','Model','Sale Price']] = 'background-color: #FFC0CB'
    return df

#sorting the data in descending order based on the sale price
sorted_data1 = car_data.sort_values(by="Sale Price",ascending=False,ignore_index=True)

#displaying the top 5 cars' data by highlighting the columns 'Make','Model',and 'Sale Price'
sorted_data1.head().style.apply(highlight_cols,axis=None)
```

```python
#alternative way to find the top 5 cars based on sale price without sorting
new_data=car_data.nlargest(5,'Sale Price')[['Make','Model','Sale Price']]
new_data=new_data.set_index(pd.Index([1,2,3,4,5]))
new_data
```

### 5. Find the least 5 cars based on their Resell price.

```python
#function to give a background colour
def highlight_cols(x):
    df=x.copy()
    df.loc[:, :] = 'background-color: #EEEEE0'
    df[['Make','Model','Resell Price']] = 'background-color: #FFC0CB'
    return df

#sorting the data in ascending order based on the Resell price
sorted_data2 = car_data.sort_values(by="Resell Price",ignore_index=True)

#displaying least 5 cars in the sorted data by highlighting the columns 'Make','Model',and 'Resell'
sorted_data2.head().style.apply(highlight_cols,axis=None)
```

```python
#alternative way to display least 5 cars based on Resell price
new_data = car_data.nsmallest(5,'Resell Price')[['Make','Model','Resell Price']]
new_data = new_data.set_index(pd.Index([1,2,3,4,5]))
new_data
```
