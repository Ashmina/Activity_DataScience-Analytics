### 1.Reading the dataset to python environment

```python
iris_data = pd.read_excel('iris.xls')
```

```python
iris_data.columns
```

The iris dataset is having 5 columns, namely, SL(sepal length), SW(sepal width), PL(petal length), PW(petal width), and Classification.

### 5.Visualizations 

#### 5.1. Count plot

```python
#Count plot
sns.countplot(x='Classification',data=iris_data)
plt.title("Count plot of Classification",fontsize=15)
plt.show()
```

The dataset contains features of three species of Iris flower (Iris-setosa, Iris-versicolor, Iris-virginica). The total rows is 150 and each class is having 50 records.

#### 5.2. Box plot

```python
#Box plot
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='PL',data=iris_data)
plt.title("Petal Length",fontsize=15)

plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='PW',data=iris_data)
plt.title("Petal Width",fontsize=15)

plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='SL',data=iris_data)
plt.title("Sepal Length",fontsize=15)

plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='SW',data=iris_data)
plt.title("Sepal Width",fontsize=15)

plt.tight_layout()
```

The values of four features, except Sepal Width(SW), are maximum for the species Iris-virginica and minimum for the species Iris-setosa. Sepal width is maximum in Iris-setosa and minimum in Iris-versicolor.

#### 5.3. Scatter plots

```python
#Scatter plots of pair of 4 features(petal length,petal width,sepal length,sepal width) grouped by Classification(Iris-setosa, Iris-versicolor, Iris-verginica)
plt.figure(figsize=(18,15))

#Scatter plot on petal length and petal width
plt.subplot(3,2,1)
sns.scatterplot(x= 'PL', y='PW',hue = 'Classification',  data = iris_data, s=50)

#Scatter plot on petal length and sepal length
plt.subplot(3,2,2)
sns.scatterplot(x= 'PL', y='SL',hue = 'Classification',  data = iris_data, s=50)

#Scatter plot on petal length and sepal width
plt.subplot(3,2,3)
sns.scatterplot(x= 'PL', y='SW',hue = 'Classification',  data = iris_data, s=50)

#Scatter plot on sepal length and sepal width
plt.subplot(3,2,4)
sns.scatterplot(x= 'SL', y='SW',hue = 'Classification',  data = iris_data, s=50)

#Scatter plot on sepal length and petal width
plt.subplot(3,2,5)
sns.scatterplot(x= 'SL', y='PW',hue = 'Classification',  data = iris_data, s=50)

#Scatter plot on petal width and sepal width
plt.subplot(3,2,6)
sns.scatterplot(x= 'PW', y='SW',hue = 'Classification',  data = iris_data, s=50)

plt.tight_layout()
```

From the above scatter plots it is clear that the species Iris-setosa can be easily differentiated using any of the four features, while classifying the other two species requires some more analysis.

#### 5.4. Heatmap

```python
#copying iris_data to a new dataframe
new_iris_data = iris_data.copy()
#to create a heatmap,data should be in numerical form
#replacing string values in the column 'Classification' with integer values(1 for Iris-setosa,2 for Iris-versicolor,3 for Iris-virginica)
new_iris_data["Classification"] = new_iris_data['Classification'].replace(['Iris-setosa','Iris-versicolor','Iris-virginica'],[1,2,3])

#heatmap
plt.figure(figsize = (8,8) )
plt.title('Heatmap on Correlation of features in the data',fontsize=16,pad=20)
sns.heatmap(new_iris_data.corr(),annot = True)
plt.xticks(rotation = 90)
plt.show()
```

The features PL(petal length), PW(petal width), and SL(sepal length) are positively correlated to Classification. Among that PL and PW have strong dependancy with Classification. The feature SW(sepal width) has inverse dependancy with Classification.
